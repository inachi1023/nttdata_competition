```python
#画像ファイル（npyファイル）を読み込み
load_East_China_all = np.load('NTL-dataset/npy/East_China.npy')
load_France_all = np.load('NTL-dataset/npy/France.npy')
load_Italy_all = np.load('NTL-dataset/npy/Italy.npy')
load_USA_all = np.load('NTL-dataset/npy/USA.npy')
load_West_China_all = np.load('NTL-dataset/npy/West_China.npy')
```

```python
#各国の画像ファイルに対して画素最大値255で割って、その合計値を配列に格納
East_China_all = []
France_all = []
Italy_all = []
USA_all = []
West_China_all = []

for  i in range(15):
    l = load_East_China_all[i, :, :] 
    j = l / 255
    n = round(j.sum())
    East_China_all.append(n)
    l = []

for  i in range(15):
    l = load_France_all[i, :, :] 
    j = l / 255
    n = round(j.sum())
    France_all.append(n)
    l = []

for  i in range(15):
    l = load_Italy_all[i, :, :] 
    j = l / 255
    n = j.sum()
    Italy_all.append(n)
    l = []
    
for  i in range(15):
    l = load_USA_all[i, :, :] 
    j = l / 255
    n = round(j.sum())
    USA_all.append(n)
    l = []
    
for  i in range(15):
    l = load_West_China_all[i, :, :] 
    j = l / 255
    n =round( j.sum())
    West_China_all.append(n)
    l = []
```
